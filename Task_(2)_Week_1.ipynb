## 1. Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/content/Datasets/Task (2) Dataset.csv')
df
```

### Identify and analyze patterns in the data.

```python
df.hist(figsize=(12, 8))
plt.show()
```

```python
df.info()
```

### Visualize the distribution of key variables (e.g., Traffic_Index, Date).

```python
df.plot.scatter(x= 'JamsCount', y='JamsDelay')
```

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
plt.figure(figsize=(8, 6))
df['Datetime'].dt.date.value_counts().plot(kind='bar')
plt.title('Distribution of Datetime')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['TrafficIndexLive'].hist(bins=20)
plt.title('Distribution of TrafficIndexLive')
plt.xlabel('TrafficIndexLive')
plt.ylabel('Frequency')
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
df.plot.hexbin(x='JamsCount',y='TrafficIndexWeekAgo', gridsize=10)
```
